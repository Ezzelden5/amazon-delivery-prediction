# ⚛ data cleaning

## ▶ solving data problems

```python
# from indepth info we can say this data has 7 issues
# 1- [Traffic, Vehicle, Area] columns has white space in their string values
#     --: df['Traffic'] = df['Traffic'].str.strip()
# =====================================================================
# 2- in [Traffic] column has 'NaN' as string not as null value
#     --: df['Traffic'] = df['Traffic'].replace('NaN', np.nan) -> then fill it
# ==========================================================================================
# 3- a missing values in [Agent_Rating, Weather] columns Agent_Rating = 0.12% & Weather has 0.21%
#     --: Fill Agent_Rating with median, fill Weather with mode
# ===================================================================================================
# 4- in [Agent_Rating] column there is 53 rows has rate 6.0 which means it was out of range (1-5) & also has only 38 of 1.0 ratings that are suspicious
#     --: Cap values > 5.0 (treat as outlier) or remove those rows
# =====================================================================================================
# 5- in [Agent_Ages] column there is 38 rows with 15 years old (under age 18) there is no hiring under the age
#     --: Remove or replace rows where Agent_Age < 18
# ========================================================================================================
# 6- there is a probarly problem with drop & pick long & latt columns will see it in the feature engneering phase
#     --: Remove rows where calculated distance > a reasonable threshold
# ==============================================================================================================
# 7- there is a probarly problem with pick up time & delivary time columns will see it in the feature engneering phase
```

### ➖ problem 1

```python
df['Order_Time'].unique()
print(df['Area'].unique())
print(df['Traffic'].unique())
print(df['Vehicle'].unique())
```

```python
object_columns = df.select_dtypes(include='object').columns

for col in object_columns:

    original_values = df[col].astype(str)

    df[col] = original_values.str.strip()


```

```python
df['Order_Time'].unique()
print(df['Area'].unique())
print(df['Traffic'].unique())
print(df['Vehicle'].unique())
```

### ➖ problem 2

```python
nan_strings = ['NaN', 'nan', 'NULL', 'null', 'None', '']

for col in df.select_dtypes(include='object').columns:

    original_series = df[col].copy()

    df[col] = df[col].replace(nan_strings, np.nan)


```

```python
(original_series.notna() & df[col].isna()).sum()
```

### ➖ problem 3

handling agent_rating missing values by median

```python
missing_count = df['Agent_Rating'].isna().sum()
if missing_count > 0:
    median_agent_rating = df['Agent_Rating'].median()
    df['Agent_Rating'] = df['Agent_Rating'].fillna(median_agent_rating)
```

```python
df['Agent_Rating'].isna().mean()
```

handling weather missing values

```python
missing_weather = df['Weather'].isna().sum()
if missing_weather > 0:
    mode_weather = df['Weather'].mode()[0]
    df['Weather'].fillna(mode_weather, inplace=True)
```

```python
df['Weather'].isna().mean()
```

handling Traffic missing values

```python
missing_traffic = df['Traffic'].isna().sum()
if missing_traffic > 0:
    mode_traffic = df['Traffic'].mode()[0]
    df['Traffic'].fillna(mode_traffic, inplace=True)
```

```python
df['Traffic'].isna().mean()
```

### ➖ problem 4

handling Agent_Rating outlier rating (out of range rating)

```python
outlier_rating = (df['Agent_Rating'] > 5.0).sum()
if outlier_rating > 0:
    df.loc[df['Agent_Rating'] > 5.0, 'Agent_Rating'] = 5.0
```

```python
df['Agent_Rating'].value_counts()
```

### ➖ problem 5

handling Agent_Age outlier age (out of range age)

```python
suspicious_age = (df['Agent_Age'] < 18).sum()
if suspicious_age > 0:
    df = df[df['Agent_Age'] >= 18].copy()
```

```python
df['Agent_Age'].value_counts()
```

### ➖ problem 6

handling datetime data type for date columns

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Order_Time_parsed'] = pd.to_datetime(df['Order_Time'], format='%H:%M:%S', errors='coerce')
df['Pickup_Time_parsed'] = pd.to_datetime(df['Pickup_Time'], format='%H:%M:%S', errors='coerce')
```

```python
df['Store_Point'] = df['Store_Latitude'].astype('str') + ', ' + df['Store_Longitude'].astype('str')
df['Drop_Point'] = df['Drop_Latitude'].astype('str') + ', ' + df['Drop_Longitude'].astype('str')
df.head(1)
```

```python
from geopy.distance import great_circle

def distance_fun(x):
  return round(great_circle(x['Store_Point'], x['Drop_Point']).kilometers, 2)
```

```python
df['distance_km'] = df.apply(distance_fun, axis=1)
df.head(1)
```

### ➖ feature columns

```python
df_ml = df.drop(columns=['Order_ID', 'Order_Date', 'Order_Time', 'Pickup_Time', 'Order_Time_parsed', 'Pickup_Time_parsed', 'Store_Latitude', 'Store_Longitude', 'Drop_Latitude', 'Drop_Longitude', 'Store_Point', 'Drop_Point'], axis=1)
df_ml.head()
```

```python
X = df_ml.drop('Delivery_Time', axis=1)
y = df_ml['Delivery_Time']
```

### ➖ OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline

ord_encod_traffic = ['Low', 'Medium', 'High', 'Jam']
ord_encod_weather = ['Sunny', 'Windy', 'Sandstorms', 'Cloudy', 'Fog', 'Stormy']
ord_encod_vehicle = ['bicycle', 'scooter', 'motorcycle', 'van']
ord_encod_area = ['Other', 'Semi-Urban', 'Urban', 'Metropolitian']

ordinal_col = ['Traffic', 'Weather', 'Vehicle', 'Area']

ord_pipeline = Pipeline(steps=[('ord_encod_traffic', OrdinalEncoder(categories=[ord_encod_traffic, ord_encod_weather, ord_encod_vehicle, ord_encod_area]))])
ord_pipeline
```

### ➖ OneHOtEncoder

```python
from sklearn.preprocessing import OneHotEncoder

ohe_col = ['Category']
ohe = OneHotEncoder(handle_unknown='ignore')
ohe_pipeline = Pipeline(steps=[('ohe', ohe)])
ohe_pipeline
```

### ➖ Standared scaler

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

scaler_col = X.select_dtypes(include='number').columns

scaler_pip = Pipeline(steps=[('scaler', StandardScaler())])
scaler_pip
```

### ➖ Column Transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder


preprocessing = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), scaler_col),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ohe_col),
        ('ord', OrdinalEncoder(categories=[ord_encod_traffic, ord_encod_weather, ord_encod_vehicle, ord_encod_area]), ordinal_col)
    ],
    remainder='passthrough'
)
preprocessing
```

# ⚛ Model Test

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
# from sklearn.preprocessing import PolynomialFeatures
# from sklearn.neighbors import KNeighborsRegressor
# from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_validate
from sklearn.pipeline import Pipeline


models = [ ('Linear Regression' , LinearRegression()),
          #  ('Knn', KNeighborsRegressor()),
          #  ('SVR Linear', SVR(kernel= 'linear')),
          #  ('SVR RBF', SVR(kernel= 'rbf')),
           ('Lasso', Lasso(random_state = 42)),
           ('Ridge', Ridge(random_state= 42)),
           ('ElasticNet', ElasticNet(random_state= 42)),
          #  ('Polynomial', PolynomialFeatures(degree = 2)),
           ('Decision Tree', DecisionTreeRegressor(max_depth=10, min_samples_split=10, random_state=42)),
           ('Random Forest', RandomForestRegressor(n_estimators=150, max_depth=15, random_state=42)),
           ('XGBoost', XGBRegressor(n_estimators=150, max_depth=7, learning_rate=0.1, random_state=42)),
           ('LGBM', LGBMRegressor(random_state= 42)),
           ('CatBoost', CatBoostRegressor(random_seed = 42, verbose=0))
         ]


for model in models:
    model_pipeline = Pipeline(steps= [('Preprocessing', preprocessing), ('Model', model[1])])

    # if model[0] == 'Polynomial':
    #   model_pipeline = Pipeline(steps= [ ('Preprocessing', preprocessing_frieght), ('Polynomial', model[1]), ('Model', Ridge())])


    cv_results = cross_validate(estimator= model_pipeline, X= X, y= y, cv= 5, scoring=['r2', 'neg_mean_squared_error', 'neg_mean_absolute_error'], return_train_score= True)

    print(model[0])
    print('Train R2 Score :', round(cv_results['train_r2'].mean() * 100, 3))
    print('Test R2 Score :', round(cv_results['test_r2'].mean() * 100, 3), '\n')

    print('Train Neg Mean Absolute Error :', round(cv_results['train_neg_mean_absolute_error'].mean(), 3))
    print('Test Neg Mean Absolute Error :', round(cv_results['test_neg_mean_absolute_error'].mean(), 3))

    print('-' * 50)
```

### ➖ randomize search

```python
import joblib
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline



xgb_pipeline_rs = Pipeline(steps=[
    ('Preprocessing', preprocessing),
    ('Model', XGBRegressor(random_state=42))
])


param_grid_xgb_rs = {
    'Model__n_estimators': [100, 200, 300],
    'Model__max_depth': [5, 7, 9],
    'Model__learning_rate': [0.01, 0.05, 0.1],
    'Model__subsample': [0.8, 1.0]
}


xgb_rs = RandomizedSearchCV(
    estimator=xgb_pipeline_rs,
    param_distributions=param_grid_xgb_rs,
    n_iter=10,
    cv=5,
    scoring='r2',
    random_state=42
)



xgb_rs.fit(X, y)

print(f" Parameters: {xgb_rs.best_params_}")
print(f" best r2 score: {round(xgb_rs.best_score_ * 100, 3)}")


best_final_pipeline = xgb_rs.best_estimator_
```
